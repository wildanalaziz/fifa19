## Visualization

```python
plt.title("Players Age", fontdict={"fontsize":15})
plt.hist(data["Age"])
plt.axvline(data["Age"].mean(), color="g")
```

```python
plt.title("Oldest Players")
oldest_players = data[["Name", "Age"]].sort_values(by="Age")
plt.barh(y=oldest_players.Name[-5:], width=oldest_players.Age[-5:])
```

```python
plt.title("Youngest Players")
oldest_players = data[["Name", "Age"]].sort_values(by="Age")
plt.barh(y=oldest_players.Name[:5], width=oldest_players.Age[:5])
```

```python
plt.title("Players Nationality", fontdict={'fontsize':15})
nationality_counts = data["Nationality"].value_counts()
nationality_counts = nationality_counts[:5].sort_values()
plt.barh(y=nationality_counts.index, width=nationality_counts.values)
```

```python
club_counts=data["Club"].value_counts()
plt.title("Number of Players in Club", fontdict={'fontsize':15})
plt.barh(y=club_counts.index[:15], width=club_counts.values[:15])
```

```python
kaki_dominan = data["Preferred Foot"].value_counts()
plt.title("Kaki Dominan", fontdict={'fontsize':15})
plt.bar(x=kaki_dominan.index, height=kaki_dominan.values)
```

```python
top_players = data.sort_values(by='Overall', ascending=False)
top_players = top_players[:5].sort_values(by="Overall")
plt.barh(y=top_players.Name, width=top_players.Overall)
```
